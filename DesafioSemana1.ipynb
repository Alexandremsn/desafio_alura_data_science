### Entender quais informações o conjunto de dados possui;


Importando as Bibiotecas

```python
import pandas as pd
```

Importando os dados

```python
dataset = pd.read_json("https://raw.githubusercontent.com/sthemonica/alura-voz/main/Dados/Telco-Customer-Churn.json")
```

varificando se a importação foi bem sucedida

```python
dataset.head()
```

Verificamos que os dados possuem diversas informações dos clientes seu perfil demográfico, seus serviços contratados, e seu gasto mensal.

###Analisar quais os tipos de dados

verificamos que os dados estão agrupado em dicionários assim vamos aplicar a normallização as variáveis que estão agrupadas normalizando cada coluna individualmente e verificando sua saída.

```python
dataset_customer = pd.json_normalize(data=dataset.customer)
dataset_customer
```

Agora os dados estão normalizados em outro dataframe vamos repetir o processo para as outras colunas

```python
dataset_phone = pd.json_normalize(data=dataset.phone)
dataset_phone
```

Normalizando dados para a variável internet

```python
dataset_internet = pd.json_normalize(data=dataset.internet)
dataset_internet
```

Normalizando a varível account

```python
dataset_account = pd.json_normalize(data=dataset.account)
dataset_account
```

Juntando as variáveis normalizadas no dataset principal

```python
dataset = pd.concat([dataset, dataset_customer,dataset_phone, dataset_internet, dataset_account], axis=1)
dataset.head(10)
```

procedemos a exclusão das colunas fonte de dados
Alguns métodos do pandas podem apresentar erro se forem aplicadas em um coluna com '.' no nome da coluna, assim vamos renomear para evitar problemas


```python
del dataset['customer']
del dataset['phone']
del dataset['internet']
del dataset['account']
dataset.rename(columns={'Charges.Monthly' : 'Charges_Monthly', 'Charges.Total' : 'Charges_Total'}, inplace = True)
dataset.head()
```

O dataset foi inteiramente normalizado e agora podemos ver com clareza os dados apresentados, o dataset é predominantemente qualitativo, tendo aglumas variáveis quantitativas.

###Corrigir as inconsistências nos dados;

As inconsitências detectadas são:
Churn com valores inválidos ' ';
SeniorCitzen com valores fora de padâo;
Charges_Total com dados de outro tipo que não numérico

trocamos os valores inválidos por 0, porém vamos verificar se este espaço vazio é aunsência de dados por erro ou deve ser subtituido por não.

```python
for i in range(dataset.shape[0]):
  if (dataset.Charges_Total[i] == " "):
    dataset.Charges_Total[i] = 0
    
```

recebemos a infromação que os valores vazios significavam ausência de dados, assim vamos verificar o percentual destas inconsistências. para verificar o impacto da exclusão destas linhas na base de dados.

```python
dataset.Churn.value_counts(normalize = True)*100
```

Verificamos que 3% dos dados estão sem valor, vamos remover os valores da base pois o valor será o foco de nossa análise.

```python
print(dataset.shape)
for i in range(dataset.shape[0]):
  valores = ['No','Yes']
  if (not(dataset.Churn[i] in valores)):
    dataset.drop(i,axis =0, inplace=True)
    
dataset.reset_index(drop=True, inplace=True)   

print(dataset.shape)
```

```python
dataset.Churn.value_counts()
```

Realizamos a exclusão dos valores e verificamos que agora temos na varível apenas os valores Yes e No.

Agora vamos corrigir a inconsitência da coluna 'Charges_Total', que possuiam dados do tipo string, vamos converte-los para número

```python
dataset['Charges_Total'] = pd.to_numeric(dataset['Charges_Total'])
dataset.head()
```

```python
dataset.Charges_Total.dtypes
```

Sucesso, agora a coluna possui apenas dados do tipo numérico.

Agora vamos padornizar os dados da variável 'SeniorCitizen', trocando o o por No e o 1 por Yes.

```python
dataset['SeniorCitizen'] = dataset['SeniorCitizen'].replace([0,1],['No','Yes'])

dataset.head(10)
```

Agora as inconsitências detectadas foram corrigidas.
