```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.style as style
pd.set_option('display.max_columns', 500)
```

importando nossa bilbiotecas

```python
dataset = pd.read_csv('https://github.com/Alexandremsn/desafio_alura_data_science/raw/main/semana1/dados_normalizados_churn.csv')   
```

importando o df para trabalho e visualizando se sua importação se deu corretamente.

```python
dataset.head()
```

testamos as variáveis pois precisaremos de variáveis dummies binárias de 0 e 1 para utilizar em nosso modelos assim transformamos as variáveis abaixo mas vizualizamos seus valores com o value_counts para esccrever o código abaixo e tranformalas

```python
dataset.Conta_Online.value_counts()
```

```python
dataset.Churn.replace({'No':0,'Yes':1},inplace=True)
dataset['Male'] = dataset.Genero.replace({'Female':0,'Male':1})
dataset.Maior_65.replace({'No':0,'Yes':1},inplace=True)
dataset.Parceiro.replace({'No':0,'Yes':1},inplace=True)
dataset.Dependentes.replace({'No':0,'Yes':1},inplace=True)
dataset.Serviço_telefone.replace({'No':0,'Yes':1},inplace=True)
dataset.Mais_de_uma_linha.replace({'No':0,'Yes':1,'No phone service':0},inplace=True)
dataset.drop('Genero',inplace=True, axis=1)
dataset.Seguraca_Online.replace({'No':0,'Yes':1,'No internet service':0},inplace=True)
dataset.Backup_Online.replace({'No':0,'Yes':1,'No internet service':0},inplace=True)
dataset.Proteção_de_dispositivo.replace({'No':0,'Yes':1,'No internet service':0},inplace=True)
dataset.Suporte_Tecnico.replace({'No':0,'Yes':1,'No internet service':0},inplace=True)
dataset.Streaming_TV.replace({'No':0,'Yes':1,'No internet service':0},inplace=True)
dataset.Streaming_Filmes.replace({'No':0,'Yes':1,'No internet service':0},inplace=True)
dataset.Conta_Online.replace({'No':0,'Yes':1},inplace=True)
```

estas variáveis tinham mais opçoes categóricas que sim e não assim crimaos variáveis dummies e juntamos com o df principal e excluímos as colunas originais

```python
df_dummies = dataset[['Serviço_Internet']]
df_dummies['Tipo_Contrato'] = dataset.Tipo_Contrato
df_dummies['Metodo_de_Pagamento'] = dataset.Metodo_de_Pagamento

dataset_dummies = pd.get_dummies(df_dummies)
dataset = pd.concat([dataset,dataset_dummies],axis=1)

dataset.drop('Serviço_Internet',inplace=True, axis=1)
dataset.drop('Tipo_Contrato',inplace=True, axis=1)
dataset.drop('Metodo_de_Pagamento',inplace=True, axis=1)

dataset.head()
```

```python
dataset_dummies.head()
```

```python
dataset.info()
```

```python
verificamos que o tipo estava diferente então no código abaixo padronizamos o tipo das variáveis binárias.
```

```python
dataset['Serviço_Internet_DSL']= dataset['Serviço_Internet_DSL'].astype('int64')
dataset['Serviço_Internet_Fiber optic']= dataset['Serviço_Internet_Fiber optic'].astype('int64')
dataset['Tipo_Contrato_Month-to-month']= dataset['Tipo_Contrato_Month-to-month'].astype('int64')
dataset['Tipo_Contrato_One year']= dataset['Tipo_Contrato_One year'].astype('int64')
dataset['Tipo_Contrato_Two year']= dataset['Tipo_Contrato_Two year'].astype('int64')
dataset['Metodo_de_Pagamento_Bank transfer (automatic)']= dataset['Metodo_de_Pagamento_Bank transfer (automatic)'].astype('int64')
dataset['Metodo_de_Pagamento_Credit card (automatic)']= dataset['Metodo_de_Pagamento_Credit card (automatic)'].astype('int64')
dataset['Metodo_de_Pagamento_Electronic check']= dataset['Metodo_de_Pagamento_Electronic check'].astype('int64')
dataset['Metodo_de_Pagamento_Mailed check']= dataset['Metodo_de_Pagamento_Mailed check'].astype('int64')
```

importando as bibiliotecas para construir nosso modelo

```python

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from catboost import CatBoostClassifier as cat
from sklearn.ensemble import RandomForestClassifier
```

```python
dataset.columns
```

```python
selection = ['Churn', 'Maior_65', 'Parceiro', 'Dependentes',
       'Meses_de_Contrato', 'Serviço_telefone', 'Mais_de_uma_linha',
       'Seguraca_Online', 'Backup_Online', 'Proteção_de_dispositivo',
       'Suporte_Tecnico', 'Streaming_TV', 'Streaming_Filmes', 'Conta_Online',
       'Gasto_Diario', 'Gasto_Mensal', 'Gasto_Total', 'Male',
       'Serviço_Internet_DSL', 'Serviço_Internet_Fiber optic',
       'Serviço_Internet_No', 'Tipo_Contrato_Month-to-month',
       'Tipo_Contrato_One year', 'Tipo_Contrato_Two year',
       'Metodo_de_Pagamento_Bank transfer (automatic)',
       'Metodo_de_Pagamento_Credit card (automatic)',
       'Metodo_de_Pagamento_Electronic check',
       'Metodo_de_Pagamento_Mailed check']
dataset_fit=dataset[selection]
```

criamos uma selecção primária com todas as variáveis e depois dividimos os dados em treino e teste

```python
X_train, X_test, y_train, y_test = train_test_split(dataset_fit.drop('Churn',axis=1),dataset.Churn,test_size=0.2, random_state=42)
```

```python
model_neighbors = KNeighborsClassifier(n_neighbors=6)
model_neighbors.fit(X_train,y_train)
y_pred_KNC = model_neighbors.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, recall_score, f1_score
```

criamos um modelo simples para teste e importamos as bibliotécas de métricas e implementamos uma função para avaliar nossos modelos.

```python
def avalia_modelo(y_pred,y_test):
  accuracy = accuracy_score(y_test,y_pred)
  recall = recall_score(y_test,y_pred)
  f1 = f1_score(y_test,y_pred)

  print('accuracy is :', accuracy)
  print('recall is :', recall)
  print('f1 score is :', f1)
```

```python
avalia_modelo(y_pred_KNC,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

agora vamos usar grades de parâmetros para testar variações de hiperparâmetros.

```python
list_steps = np.linspace(2, 60, 20, dtype=int)
parameters = {'n_neighbors':list_steps}
model_KNN= KNeighborsClassifier()
model_neighbors_otimized = GridSearchCV(estimator=model_KNN,param_grid = parameters,scoring='f1',cv=20)
model_neighbors_otimized.fit(dataset_fit.drop('Churn',axis=1),dataset.Churn)
y_pred_KNC_better = model_neighbors_otimized.predict(X_test)
```

testamos o nosso modelo e verificamos a melhor configuração de modelo para nosso primeiro modelo e previmos as variável alvo com esta configuração.

```python
pd.DataFrame(model_neighbors_otimized.cv_results_)
```

```python
avalia_modelo(y_pred_KNC_better,y_test)
```

tivemos uma melhora significativa depois testamos a random forest, utilizamos configurações mais conservadoras de hiperparâmetros, pois estou desenvolvendo este notebook para fins educacionais, e apesar de melhor resultado potencial o tempo computacional seria muito extenso, assim optei por usar esta configuração neste o no próximo modelo.

```python
model_forest = RandomForestClassifier()
n_depth = np.linspace(5, 8, 3, dtype=int)
n_iterations = [500,1000,2000]
parameters = {'n_estimators':n_iterations,'max_depth':n_depth}
model_forest_otimized = GridSearchCV(estimator=model_forest,param_grid = parameters,scoring='f1',cv=2)
model_forest_otimized.fit(dataset_fit.drop('Churn',axis=1),dataset.Churn)
y_pred_model_forest = model_forest_otimized.predict(X_test)
```

```python
avalia_modelo(y_pred_model_forest,y_test)
```

```python
from catboost.core import CatBoostClassifier
model_cat = CatBoostClassifier()
n_depth = np.linspace(5, 8, 3, dtype=int)
n_iterations = [500,1000,2000]
parameters = {'iterations':n_iterations,'depth':n_depth}
model_cat_otimized = GridSearchCV(estimator=model_cat,param_grid = parameters,scoring='f1',cv=2)
model_cat_otimized.fit(dataset_fit.drop('Churn',axis=1),dataset.Churn)
y_pred_model_cat = model_cat_otimized.predict(X_test)
```

assim o modelo de boost teve um desempenho melhor vamos aprofundar os testes usando este modelo.

```python
avalia_modelo(y_pred_model_cat,y_test)
```
